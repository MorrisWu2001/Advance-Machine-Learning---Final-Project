```python
# Exploratory Data Analysis with Binned Features

# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# Load the dataset
df = pd.read_csv("diabetes_dataset.csv")

# Drop any unnamed index column if present
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]

# Display basic information
print("Dataset Info:")
df.info()
print("\nSummary Statistics:")
display(df.describe())

# Check for missing values
print("\nMissing Values:")
display(df.isnull().sum())

# Display first few rows
print("\nFirst Few Rows:")
display(df.head())

# Identify numeric columns
numeric_cols = df.select_dtypes(include=np.number).columns.tolist()
print("\nNumeric columns to be binned:", numeric_cols)

# Create binned versions of each numeric column using quartiles and equal-width bins
for col in numeric_cols:
    try:
        df[f'{col}_quartile_bin'] = pd.qcut(df[col], q=4, labels=["Q1", "Q2", "Q3", "Q4"], duplicates='drop')
    except ValueError as e:
        print(f"Skipping quartile binning for {col}: {e}")
    try:
        df[f'{col}_equal_bin'] = pd.cut(df[col], bins=4, labels=["Bin1", "Bin2", "Bin3", "Bin4"])
    except ValueError as e:
        print(f"Skipping equal-width binning for {col}: {e}")

# Display a sample of the new binned columns
binned_columns = [col for col in df.columns if '_bin' in col]
print("\nSample of binned columns:")
display(df[binned_columns].head())

# Plot the distribution of binned features
for col in binned_columns:
    plt.figure(figsize=(6, 4))
    sns.countplot(x=col, data=df, order=df[col].value_counts().index)
    plt.title(f'Distribution of {col}')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Analyze relationship between binned features and outcome variable if present
outcome_col = 'targets' if 'targets' in df.columns else 'Outcome'
if outcome_col in df.columns:
    for col in binned_columns:
        plt.figure(figsize=(6, 4))
        sns.countplot(x=col, hue=outcome_col, data=df)
        plt.title(f'{col} vs {outcome_col}')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()

# Correlation heatmap of numerical features
plt.figure(figsize=(12, 10))
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Numeric Features')
plt.show()

# Optional: Pairplot for selected numeric features
selected_cols = numeric_cols[:4] + [outcome_col] if outcome_col in df.columns else numeric_cols[:4]
if len(selected_cols) > 1:
    sns.pairplot(df[selected_cols], hue=outcome_col if outcome_col in df.columns else None)
    plt.suptitle("Pairplot of Selected Features", y=1.02)
    plt.show()

# One-hot encode the binned columns
df_encoded = pd.get_dummies(df, columns=binned_columns, drop_first=True)

# Display the shape of the encoded DataFrame
print("\nShape after encoding:", df_encoded.shape)
```
